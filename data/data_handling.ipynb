```python
import pandas as pd

price_raw = pd.read_excel('Daily물가지수.xlsx', sheet_name='p', skiprows=13, index_col=0).iloc[6:]
price_raw.index = pd.to_datetime(price_raw.index)
price_raw

# 칼럼명에 -US 접두사 다 지우기
price_raw.columns = price_raw.columns.str.replace('-US', '')
price_raw.to_csv('price_raw.csv')
```

```python
price_raw.columns
```

```python
# yfinance 로 주가 데이터 가져오기, price_raw 칼럼 이용
import yfinance as yf
stock_data = yf.download(list(price_raw.columns), start='2000-01-01', end='2025-03-24')

# BRK.B, BF.B 는 BRK-B, BF-B 로 변경되었으므로 다시 다운로드 하여 병합

temp_data = yf.download(['BRK-B', 'BF-B'], start='2000-01-01', end='2025-03-24')
temp_data

# 날짜 기준으로 병합하고 종가만 남기기
stock_data = pd.concat([stock_data, temp_data], axis=1)
stock_data = stock_data[['Close']]
stock_data
```

```python
stock_data.to_csv('stock_data.csv')
```

```python
stock_data = pd.read_csv('stock_data.csv', index_col=0, parse_dates=True)
stock_data

stock_data_return = stock_data.pct_change()
stock_data_return

stock_data_return.to_csv('stock_data_return.csv')






```

```python
cpi_data = pd.read_excel('CPI_dataset.xlsx', index_col=0, parse_dates=True)
cpi_data

# 섹터들 값 없는 행 드롭
cpi_data = cpi_data[cpi_data.notna().all(axis=1)]
cpi_data
```

```python
sector_index_tickers = [
    "^SP500-55",  # S&P 500 Utilities Sector
    "^SP500-40",  # S&P 500 Financials Sector
    "^SP500-25",  # S&P 500 Consumer Discretionary Sector
    "^SP500-20",  # S&P 500 Industrials Sector
    "^SP500-15",  # S&P 500 Materials Sector
    "^SP500-1010",  # S&P 500 Energy Sector
    "^SP500-30",  # S&P 500 Consumer Staples Sector
    "^SP500-35",  # S&P 500 Health Care Sector
    "^SP500-45",  # S&P 500 Information Technology Sector
    "^SP500-50",  # S&P 500 Communication Services Sector
    "^SP500-60",  # S&P 500 Real Estate Sector
]

import yfinance as yf

# 섹터들의 주가 데이터 가져오기
sector_indices = yf.download(sector_index_tickers, start='2000-01-01', end='2025-03-24')
sector_indices

# 주가 데이터 정리
sector_indices = sector_indices['Close']
sector_indices.to_csv('sector_indices.csv')
sector_indices
```

```python
sector_indices_return = sector_indices.pct_change()
sector_indices_return.to_csv('sector_indices_return.csv')
```

```python
# sector_indices랑 cpi_data 합치기
cpi_data = pd.concat([cpi_data, sector_indices], axis=1)
cpi_data

# cpi_data 저장
cpi_data.to_csv('cpi_data.csv')

```
